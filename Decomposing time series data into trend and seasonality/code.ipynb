```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load the dataset
file_path = "/content/dataset.csv"
df = pd.read_csv(file_path)

# Convert 'Date' to datetime format and set it as index
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
df.set_index('Date', inplace=True)

# Perform decomposition on the 'Close' price
result = seasonal_decompose(df['Close'], model='additive', period=30)  # Assuming monthly seasonality

# Plot the decomposition
plt.figure(figsize=(10, 8))
plt.subplot(411)
plt.plot(df['Close'], label='Original Data')
plt.legend()

plt.subplot(412)
plt.plot(result.trend, label='Trend', color='red')
plt.legend()

plt.subplot(413)
plt.plot(result.seasonal, label='Seasonality', color='green')
plt.legend()

plt.subplot(414)
plt.plot(result.resid, label='Residuals', color='blue')
plt.legend()

plt.tight_layout()
plt.show()
```
